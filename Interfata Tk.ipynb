```python
import cv2
import numpy as np
from tkinter import*
import tkinter as tk
from PIL import Image, ImageTk

#--------Definire parametri-------
kernelSize = 5

#---------------------------------

#----------definirea functiilor-------------
def Pornire_Preprocesare():
    cap_cam = cv2.VideoCapture(0)

    while True:
        _, img = cap_cam.read()
        # De aici se scrie cod
        imagine_gri = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
        imagine_blur_gaussian = cv2.GaussianBlur(imagine_gri, (kernelSize, kernelSize), 0)
        margini_detectate = cv2.Canny(imagine_blur_gaussian, 50, 150)  # 1:3 ratie
        # si pana aici se va scrie cod

        # Aici se afiseaza imaginea camerelor
        cv2.imshow('Camera Pornită', img)  # afisare imagine camera originala
        cv2.imshow('Camera Convertita in gri', imagine_gri)  # afisare imagine camera originala
        cv2.imshow('Camera cu blur gausian', imagine_blur_gaussian)  # afisare imaginea camerei web cu blur gausian
        cv2.imshow('Camera cu margini detectate', margini_detectate)  # afisare imagine camera originala

        # pana aici
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap_cam.release()
    cv2.destroyAllWindows()
#-------------------------------------------

root = Tk()

var1 = IntVar()#doar aici pot trece var1 si var2 si nu inteleg de ce
var2 = IntVar()

#------------------Cadrul 1 Preprocesare imagine----------------------
Cadrul1 = LabelFrame(root,text="Preprocesare imagine")
Cadrul1.pack(fill="both", expand="yes", padx=20, pady=20)

labelc1_io = Label(Cadrul1, text="Imagine originală")
labelc1_io.pack(side=tk.LEFT, padx=10,pady=10)

labelc1_ig = Label(Cadrul1, text="Imagine gri")
labelc1_ig.pack(side=tk.LEFT, padx=10,pady=10)

labelc1_ibg = Label(Cadrul1, text="Imagine Blur Gaussian")
labelc1_ibg.pack(side=tk.LEFT, padx=10,pady=10)

labelc1_imd = Label(Cadrul1, text="Imagine margini detectate")
labelc1_imd.pack(side=tk.LEFT, padx=10,pady=10)

btn_pc1 = Button(Cadrul1, text='Pornire Camera',command = Pornire_Preprocesare)#aici trebuie sa introduce parametrul command si sa ii ofer un parametru pentru pornire
btn_pc1.pack(side=tk.BOTTOM, padx=10, pady=10)#aici trebuie sa mut butonul in stanga

#---------------------------------------------------------------------

thresh = Scale(Cadrul1, label="thresh1", from_=0, to=255, orient=HORIZONTAL, variable=var1, activebackground='#339999')
thresh.set(0)
thresh.place(x=500, y=10, width=200)
thresh2 = Scale(Cadrul1, label="thresh2", from_=255, to=0, orient=HORIZONTAL, variable=var2, activebackground='#339999')
thresh2.set(255)
thresh2.place(x=500, y=80, width=200)

#-------------------------------Cadrul2 Transformata Hough------------
Cadrul2 = LabelFrame(root,text="Transformata Hough")
Cadrul2.pack(fill="both", expand="yes", padx=20, pady=20)
#---------------------------------------------------------------------

#-----------------------------Cadru3 Invatare Automata----------------
Cadrul3 = LabelFrame(root,text="Învățare Automată")
Cadrul3.pack(fill="both", expand="yes", padx=20, pady=20)
#---------------------------------------------------------------------

root.title("Detectare traiectorii plane")#titlul ferestrei
root.geometry("800x600")#aici alegem dimensiunea ferestrei (poate lipsi aceasta linie de cod)
root.mainloop()
```
